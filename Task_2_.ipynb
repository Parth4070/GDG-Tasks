## F1 EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("f1_dnf.csv")
df.head()
```

```python
df_copy = df.copy()
```

```python
df_copy
```

```python
df_copy.isnull().sum()
```

# Cleaning

```python
import numpy as np
```

```python
df_copy[['milliseconds', 'fastestLap', 'rank','fastestLapTime','fastestLapSpeed']] = df_copy[['milliseconds', 'fastestLap', 'rank','fastestLapTime','fastestLapSpeed']].replace('\\N', np.nan)
```

```python
df_copy.head()
```

```python
df_copy.drop(['resultId', 'raceId', 'dob'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df_copy['dob'] = df['dob']
```

```python
df_copy.head()
```

```python
df_copy['date'] = pd.to_datetime(df['date'])
df_copy['dob'] = pd.to_datetime(df['dob'])
```

```python
#Calculating drivers age at the time of race
df_copy['driver_age'] = (df_copy['date'] - df_copy['dob']).dt.days  / 365.25
df_copy['driver_age'] = df_copy['driver_age'].round()
```

```python
df_copy['driver_age'].head()
```

```python
df_copy.info()
```

```python
df_copy['target_finish'] = df_copy['target_finish'].replace(1, 'Finished')
df_copy['target_finish'] = df_copy['target_finish'].replace(0, 'Not Finished')
```

```python
df_copy['driver_name'] = df_copy['forename'] + " " + df_copy['surname']
```

```python
df_copy.drop(['forename', 'surname'], axis=1, inplace=True)
```

```python
df_copy['driver_name'].head()
```

```python

df_copy.drop(['circuitId'], axis = 1, inplace=True)
```

Dropped Columns:
resultId, raceId, forename & surname (merged to driver_name), dob,

```python
df_copy.drop(['dob'], axis=1, inplace=True)
```

```python
df_copy.drop(['fastestLap', 'rank'], axis=1, inplace=True)
```

```python
df_copy.isnull().sum()
```

```python
df
```

```python
plt.subplot(1,2,1)
sns.kdeplot(df_copy['laps'])

plt.subplot(1,2,2)
sns.boxplot(df_copy['laps'])
```

```python
df_copy['laps'] = df_copy['laps'].fillna(df_copy['laps'].median())
```

add milliseconds to the copied data and convert to int


```python
# Droppping 'milliseconds' because it contains 72% missing values
df_copy.drop(['milliseconds'], axis=1, inplace=True)
```

```python
df_copy = df_copy.rename(columns={'name_y' : 'constructor'})
```

```python
df_copy = df_copy.rename(columns={'nationality_x' : 'driver_nationality', 'nationality_y':'constuctor_nationality'})
```

```python
df_copy.columns
```

```python
df_copy.head()
```

```python
df_copy.drop(['driverRef', 'constructorRef', 'circuitRef'], axis=1, inplace=True)
```

```python
df_copy.drop(['fastestLapTime', 'fastestLapSpeed'], axis=1, inplace=True)
```

```python
df_copy['points'] = df_copy.apply(
    lambda row: 0 if pd.isna(row['points']) and row['positionOrder'] > 10 else row['points'], axis=1
)
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5)
df_copy['points'] = imputer.fit_transform(df_copy[['points']])

df_copy['points'] = df_copy['points'].round()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Number of Finished and not finished

```python
grp_counts = df_copy['target_finish'].value_counts()

plt.bar(grp_counts.index, grp_counts.values)
plt.xlabel("Car Status")
plt.ylabel("No. of Cars")
plt.title("Finished vs Not Finished")
plt.show()
```

## Graphs based on Drivers
---


# Top 20 Drivers with most Finishes

```python
df_n  =  df_copy[df_copy['target_finish'] == 'Finished'].groupby('driver_name').size().reset_index(name='Count')
df_top_20 = df_n.sort_values(by='Count', ascending=False).head(20)
df_top_20 = df_top_20.sort_values(by='Count', ascending=True)

plt.barh(df_top_20['driver_name'], df_top_20['Count'], align="edge")
plt.title("Top 20 Drivers with most Finishes")
plt.show()
```

# Top 20 Drivers with most DNFs

```python
df_n  =  df_copy[df_copy['target_finish'] == 'Not Finished'].groupby('driver_name').size().reset_index(name='Count')
df_top_20 = df_n.sort_values(by='Count', ascending=False).head(20)
df_top_20 = df_top_20.sort_values(by='Count', ascending=True)

plt.barh(df_top_20['driver_name'], df_top_20['Count'], align="edge")
plt.title("Top 20 Drivers with most DNFs")
plt.show()
```

# Performance overview of Top F1 Drivers

```python
played = df_copy.groupby(by='driver_name').size()
won = df_copy[df_copy['target_finish'] == 'Finished'].groupby(by='driver_name').size()
lost = df_copy[df_copy['target_finish'] == 'Not Finished'].groupby(by='driver_name').size()

merged = pd.DataFrame({
    'Races Participated': played,
    'Races Won': won,
    'Races Lost': lost
}).fillna(0)

merged_top_20 = merged.sort_values(by='Races Won', ascending=False).head(20)

merged_top_20.plot(kind='bar', figsize=(10,6))

plt.title('Performance Overview of Top F1 Drivers: Races Participated, Won, and Lost', fontsize=14)
plt.xlabel('Driver', fontsize=12)
plt.ylabel('No. of Matches', fontsize=12)
plt.legend(title='Target Finish')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Top 20 Drivers with most points earned

```python
df_const_points = df_copy.groupby('driver_name', as_index=False)['points'].sum()
df_const_points = df_const_points.sort_values(by='points', ascending=False).head(20)
df_const_points = df_const_points.sort_values(by='points', ascending=True)

plt.barh(df_const_points['driver_name'], df_const_points['points'], align="edge")
plt.title("Top 20 Drivers with most Points earned")
plt.show()
```

```python
df_n  =  df_copy[df_copy['target_finish'] == 'Finished'].groupby('driver_name').size().reset_index(name='Count')
df_top_20 = df_n.sort_values(by='Count', ascending=False).head(20)
df_top_20
```

# Top 20 Car Constructors

```python
total_cars = df_copy.groupby(by='constructor').size()
won = df_copy[df_copy['target_finish'] == 'Finished'].groupby(by='constructor').size()
lost = df_copy[df_copy['target_finish'] == 'Not Finished'].groupby(by='constructor').size()

merged = pd.DataFrame({
    'Total Cars': total_cars,
    'Races Won': won,
    'Races Lost': lost
}).fillna(0)

merged_top_20 = merged.sort_values(by='Races Won', ascending=False).head(20)

merged_top_20.plot(kind='bar', figsize=(10,6))

plt.title('Performance Overview of Top F1 Car Constructors', fontsize=14)
plt.xlabel('Cars', fontsize=12)
plt.ylabel('No. of Races', fontsize=12)
plt.legend(title='Target Finish')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
